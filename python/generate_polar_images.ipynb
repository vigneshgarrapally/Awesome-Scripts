```
!pip install filetype
from skimage.io import imread,imshow,imsave
import numpy as np
import os
import filetype
from skimage.transform import warp_polar
from skimage.util import img_as_float,img_as_ubyte
#function to check whether a file is image or not
def checkimage(a):
    kind=filetype.guess(a)
    if kind is None:
        return -1
    return (kind.mime).find("image")
inpath="/content/Images"
outpath="/content/polarimages"
for root,dir,files in os.walk(inpath):
  for file in files:
    if (checkimage(os.path.join(root,file))>=0):
      p=os.path.join(outpath,*(root.split("/")[2:]),file)
      dirname=p.split(".")[0]
      os.makedirs(dirname,exist_ok=True)
      image=imread(os.path.join(inpath,root,file),as_gray=True)
      image=img_as_float(image)
      polar_image=warp_polar(image)
      i=0
      polar_image=img_as_ubyte(polar_image)
      imsave(os.path.join(dirname,file.split(".")[0]+"polar"+".jpg"),polar_image)
      for i in range(polar_image.shape[0]):
        polar_image=np.vstack((polar_image[-1,:].reshape(1,(polar_image.shape[1])),polar_image[0:-1,:]))
        polar_image=img_as_ubyte(polar_image)
        imsave(os.path.join(dirname,file.split(".")[0]+"polar"+str(i)+".jpg"),polar_image)
```
